### DB schema

#### schema movie
- movie.links:
    - id: pk
    - id_movie: fk movies
    - imdbId: int unique
    - tmdbId: int unique

- movie.movies:
    - id: pk
    - title, varchar(512)

- movie.tags:
    - id: pk
    - id_movie: fk movies
    - id_user: fk users
    - tag: varchar(512)
    - timestamp: timestamp 

- movie.movie_genres:
    - id: pk
    - id_movie: fk movies
    - id_genre: fk genres

#### schema rating
- rating.ratings:
    - id: pk
    - id_movie: fk movies
    - id_user: fk users
    - rating: double(3,1)
    - timestamp: timestamp

#### schema user 
- users.users:
    - id: pk

#### schema enum
- enum.genres
    - id: pk
    - name: varchar(64)
